#Measuring Play Impact

*   Big Play Def - Turnover / Sack on 3rd Down
*   Good Play Def - Tackle for a loss / Sack on 1st or 2nd down / Incomplete Pass
*   Good Play Of - Gain of 1-9 yards 
*   Big Play Of - Gain >= 10 yards
*   Huge Play Of - TD / Convert on 4th down
